```python
import hyperopt
from hyperopt import fmin, tpe, hp

import hyperopt.pyll.stochastic
import seaborn as sns

import numpy as np
```

```python
sns.set_style("darkgrid")
sns.set_context("notebook")
```

## Uniform

```python
space = hp.uniform(label='x', low=0, high=100)
sampled_space =[hyperopt.pyll.stochastic.sample(space) for _ in range(10_000)]
sampled_space[:7]
```

```python
sns.histplot(sampled_space, stat='probability')
```

## quniform
 

```python
space = hp.quniform(label='x', low=0, high=100, q=3)
sampled_space =[hyperopt.pyll.stochastic.sample(space) for _ in range(10_000)]
sampled_space[:7]
```

```python

sns.histplot(sampled_space, stat='probability',)
```

## randint

```python
space = hp.randint('x',100, )
sampled_space =[hyperopt.pyll.stochastic.sample(space) for _ in range(10_000)]
np.array(sampled_space[:7])
```

```python
sns.histplot(np.array(sampled_space), stat='probability',)
```
